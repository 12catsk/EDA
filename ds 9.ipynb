```python
mapper2={
    'Clear':0, 
    'Cloudy':1, 
    'Light rain/snow':2
}
df['weather']=df['weather'].map(mapper2)
```

```python
df['datetime'] = df['datetime']/23
```

```python
from sklearn.preprocessing import StandardScaler

to_scale= ['temp','humidity','windspeed']
for each_feature in to_scale:
    scaler = StandardScaler()
    df[each_feature] = scaler.fit_transform(df[each_feature].values.reshape(-1,1))
```
